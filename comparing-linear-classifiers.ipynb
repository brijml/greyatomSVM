## Import necessary packages

```python
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.svm import LinearSVC
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.close_figures=False
```

## Generate random training data with positive and negative samples that are linearly separable

```python
negative = []
for i in range(100):
    rn1 = (np.random.random_sample()*5)
    rn2 = (np.random.random_sample()*5)
    negative.append([rn1,rn2])
outputs = [0] * 100

positive = []
for i in range(100):
    rn1 = (np.random.random_sample()+5)*2
    rn2 = (np.random.random_sample()+5)*2
    positive.append([rn1,rn2])

inputs = negative + positive
outputs = outputs + [1]*100

xs, ys = list(zip(*inputs))
colors = []
for o in outputs:
    if o:
        colors.append('blue')
    else:
        colors.append('red')
fig, ax = plt.subplots()
ax.scatter(xs, ys, color=colors)
```

## Run Perceptron and plot the decision boundary

```python
p = Perceptron()
p.fit(inputs, outputs)

w = p.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 15)
yy = a * xx - (p.intercept_[0]) / w[1]

ax.plot(xx, yy, color='red', label='Perceptron')
ax.legend(loc='upper left')
plt.show()
```

## Run logistic regression and plot the decision boundary

```python
lr = LogisticRegression()
lr.fit(inputs, outputs)

w = lr.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 15)
yy = a * xx - (lr.intercept_[0]) / w[1]

ax.plot(xx, yy, color='magenta', label='Logistic Regression')
plt.show()
```

## Run Support Vector Machine and plot the decision boundary

```python
svc = LinearSVC()
svc.fit(inputs, outputs)

w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 15)
yy = a * xx - (svc.intercept_[0]) / w[1]

ax.plot(xx, yy, color='green', label='SVM')
ax.legend(loc='upper left')
plt.show()
```
